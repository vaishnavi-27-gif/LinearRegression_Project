```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
df=df.loc[:,['YearsExperience','Salary']]
df
```

```python
x=df["YearsExperience"].values
y=df["Salary"].values
```

```python
n=len(x)
```

```python
sum_x=np.sum(x)
sum_y=np.sum(y)
sum_xy=np.sum(x*y)
sum_x2=np.sum(x*x)
```

```python
a=(sum_y*sum_x2-sum_x*sum_xy)/(n*sum_x2-sum_x**2)
```

```python
b=(n*sum_xy-sum_x*sum_y)/(n*sum_x2-sum_x**2)
```

```python
yearsExp=2.1
salary_pred=a+b*yearsExp
print("years Experience:",yearsExp,"Salary Prediction:",salary_pred)
```

```python
y_pred=a+b*x
```

```python
sum_of_squares=np.sum((y-y_pred)**2)
sos_total=np.sum((y-np.mean(y))**2)
r2=sum_of_squares/sos_total
```

```python
mae=np.mean(np.abs(y-y_pred))
mse=np.mean((y-y_pred)**2)
rmse=np.sqrt(mse)
```

```python
print("Model Performance:")
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
print("R-squared:",r2)
```

```python
plt.title("Linear Regression:Salary vs Years of Experiences")
plt.scatter(x,y,color="blue",label="Actual Data")
plt.plot(x,y_pred,color="red",label="Regression line")
plt.xlabel("Years Experiences")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
